````python
cursor.execute(
    """
    SELECT songDetails.genre, history.listener
    FROM songDetails
    JOIN history ON songDetails.trackName = history.trackName
    JOIN users ON history.listener = users.fullname
    WHERE users.groupNumber = ?
    """,
    (1,)  
)
group1_genres = cursor.fetchall()

from collections import defaultdict
group1_genres_by_listener = defaultdict(set)
for genre, listener in group1_genres:
    group1_genres_by_listener[listener].add(genre)

# Find common genres
common_genres = set.intersection(*group1_genres_by_listener.values())
print("Common Genres for Group 1:", common_genres)

cursor.execute(
    """
    SELECT songDetails.genre, COUNT(*)
    FROM songDetails
    JOIN history ON history.trackName = songDetails.trackName
    JOIN users ON history.listener = users.fullname
    WHERE users.groupNumber = ?
    GROUP BY songDetails.genre
    ORDER BY COUNT(*) DESC
    """,
    (1,)  
)
most_common_genre = cursor.fetchone()
if most_common_genre:
    print("Most Common Genre for Group 1:", most_common_genre[0])
else:
    print("No genres found for Group 1.")
````

````python
#Which team were you able to recommend the most songs to?

cursor.execute(
    """
    SELECT DISTINCT songDetails.trackName, songDetails.artistName, songDetails.genre
    FROM songDetails
    JOIN history ON songDetails.trackName = history.trackName
    JOIN users ON history.listener = users.fullname
    WHERE users.groupNumber = ? AND songDetails.genre IN (
        SELECT DISTINCT songDetails.genre
        FROM songDetails
        JOIN history ON history.trackName = songDetails.trackName
        JOIN users ON history.listener = users.fullname
        WHERE users.groupNumber = ?
    )
    AND songDetails.trackName NOT IN (
        SELECT history.trackName
        FROM history
        JOIN users ON history.listener = users.fullname
        WHERE users.groupNumber = ?
    )
    """,
    (0, 1, 1)  # Replace 0 with your group number, 1 with the target group number
)
recommended_songs = cursor.fetchall()

# Display recommendations
import pandas as pd
recommended_songs_df = pd.DataFrame(recommended_songs, columns=["TrackName", "ArtistName", "Genre"])
print("\nRecommended Songs:")
print(recommended_songs_df)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

cursor.execute(
    """
    SELECT songDetails.genre, COUNT(*) AS genre_frequency
    FROM songDetails
    JOIN history ON songDetails.trackName = history.trackName
    JOIN users ON history.listener = users.fullname
    WHERE users.groupNumber = ?
    GROUP BY songDetails.genre
    ORDER BY genre_frequency DESC
    LIMIT 10;
    """,
    (0,) 
)

# Fetch the results
genre_data = cursor.fetchall()

# Check if data exists
if genre_data:
    # Prepare data for the plot
    genres = [row[0] for row in genre_data]
    frequencies = [row[1] for row in genre_data]

    # Set a modern style
    plt.figure(figsize=(8, 8))
    plt.pie(frequencies, labels=genres, autopct='%1.1f%%', colors=sns.color_palette("pastel"), startangle=140)
    plt.title("Top 10 Genres Proportion for Group 0", fontsize=18, weight='bold')
    plt.tight_layout()
    plt.show()
else:
    print("No data found for the specified group.")
````
